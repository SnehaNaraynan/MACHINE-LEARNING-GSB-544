## Question 1
1. Create side-by-side boxplots of the `mpg` variable by the different values of the `am` variable. What happens if you convert the `am` variable back to float values and then try to create this same plot? Explain the differences between the two plots.

```python
boxplot_am_cat = (
    ggplot(df, aes(x='am', y='mpg')) +
    geom_boxplot() +
    labs(title='Boxplot of MPG by Transmission Type (AM) when am is categorical', x='am', y='MPG')
)

boxplot_am_float = (
    ggplot(df, aes(x=df['am'].astype(float), y='mpg')) +
    geom_boxplot() +
    labs(title='Boxplot of MPG by Transmission Type (AM) when am is float', x='am', y='MPG')
)
boxplot_am_cat.show()
boxplot_am_float.show()
```

> With `am` as a category, the boxplots show `mpg` for automatic vs. manual clearly. As a float, it creates a less useful plot by treating `am` as continuous, which doesnt make sense for categories.

## Question 2
2. Create overlaid histograms of the `hp` variable for the different values of the `cyl` variable. What happens if you convert the `cyl` variable back to float values and then try to create this same plot? Explain the differences between the two plots.

```python
plot_cyl_cat = (ggplot(df, aes(x="hp", fill="cyl")) 
                + geom_histogram(alpha=0.5, position='identity', binwidth=10))


df["cyl_float"] = df["cyl"].astype(float)
plot_cyl_float = (ggplot(df, aes(x="hp", fill="cyl_float")) 
                  + geom_histogram(alpha=0.5, position='identity', binwidth=10))

plot_cyl_cat.show()
plot_cyl_float.show()
```

> **Categorical Plot**: When `cyl` is categorical, the histogram uses different colors to clearly show the distribution of horsepower (`hp`) by cylinder count, allowing easy comparison.

> **Float Plot**: Converting `cyl` to float results in a single grayscale histogram, treating `cyl` as continuous and losing the distinction between categories.

> In short, the categorical plot is more meaningful, while the float version fails to show clear groupings.

Some of these variables, like the `cyl` variable, have numeric values that actually make sense as numbers (i.e. the number of cylinders in the engine). However, it doesn't make the most sense to "do math" with this type of variable (e.g. take averages and such) because there are so few different value this can take on AND they're an explicit choice made by the car manufacturer. So, it makes more sense to treat the `cyl` variable as a categorical variable despite it having numeric values.

You will need to keep these kinds of nuances about data in mind as you work with an increasing variety and richness of data, and do more complex things with them.

### *Scatter plot of horsepower vs mpg with a trend line*

```python
scatter_plot_hp_mpg = (ggplot(df, aes(x='hp', y='mpg')) 
                       + geom_point(color="green") 
                       + geom_smooth(method='lm', color='red', se=False)
                       + labs(title="Scatter Plot of HP vs MPG", x="Horsepower", y="Miles Per Gallon"))
scatter_plot_hp_mpg.show()
```

### *Density plot of horsepower (hp)*

```python
density_plot_hp = (ggplot(df, aes(x='hp')) 
                   + geom_density(fill="blue", alpha=0.5)
                   + labs(title="Density Plot of Horsepower", x="Horsepower", y="Density"))

density_plot_hp.show()
```

### Violin plot of horsepower (hp) by cylinder count (cyl)

```python
violin_plot_hp_cyl = (ggplot(df, aes(x='factor(cyl)', y='hp', fill='factor(cyl)')) 
                      + geom_violin(alpha=0.7) 
                      + labs(title="Violin Plot of Horsepower (hp) by Cylinder Count", 
                             x="Cylinders (4, 6, 8)", 
                             y="Horsepower (hp)") 
                      + scale_fill_manual(values=["lightblue", "lightgreen", "lightcoral"])  
                      + theme(legend_title=element_blank()))  
```
